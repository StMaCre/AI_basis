```python
import re
import pandas as pd
import time

# Score extraction function (unchanged)
def extract_score_from_output(text):
    """Extracts the score from the model's text output using regex patterns that target specific score indicators."""
    if not isinstance(text, str):
        return None
    
    # Pattern 1: Look for "Score: X" or "score: X" pattern
    score_pattern = re.search(r'[Ss]core:\s*(\d)', text)
    if score_pattern and 1 <= int(score_pattern.group(1)) <= 5:
        return int(score_pattern.group(1))
    
    # Pattern 2: Look for standalone "X/5" pattern
    rating_pattern = re.search(r'(\d)/5', text)
    if rating_pattern and 1 <= int(rating_pattern.group(1)) <= 5:
        return int(rating_pattern.group(1))
    
    # Pattern 3: Look for "rating of X" or "score of X" pattern
    of_pattern = re.search(r'(rating|score) of (\d)', text, re.IGNORECASE)
    if of_pattern and 1 <= int(of_pattern.group(2)) <= 5:
        return int(of_pattern.group(2))
    
    # Pattern 4: Last resort - find the first standalone digit between 1-5 that's likely a score
    standalone_pattern = re.search(r'(?:^|\s|[.,;:])([1-5])(?:\s|[.,;:]|$)', text)
    if standalone_pattern:
        return int(standalone_pattern.group(1))
    
    return None

# Function to create few-shot examples (unchanged)
def create_few_shot_examples(df, question_cols, score_col, num_examples=10):
    """Creates formatted few-shot examples from the dataframe."""
    # Filter valid rows with non-null values in both questions and the score
    valid_examples = df.copy()
    
    # Filter rows with score
    valid_examples = valid_examples[valid_examples[score_col].notna()]
    
    # Get rows with at least the first question answered
    if len(question_cols) > 0:
        valid_examples = valid_examples[valid_examples[question_cols[0]].notna()]
    
    # Take the top examples
    valid_examples = valid_examples.head(num_examples)
    
    examples = []
    for _, row in valid_examples.iterrows():
        ex = []
        for i, col in enumerate(question_cols):
            if pd.notna(row[col]):  # Only include non-null answers
                ex.append(f"Q{i+1}: {col}\nA{i+1}: {row[col]}")
        
        ex.append(f"Score: {int(row[score_col])}")
        examples.append("\n\n".join(ex))
    
    return "\n\n---\n\n".join(examples)

# Modified function to generate prompts - handle missing data
def generate_prompts_for_scoring(rows_df, question_cols, full_prompt):
    """Generates prompts for each row using the question columns and full prompt template."""
    prompts = []
    for _, row in rows_df.iterrows():
        prompt_parts = [full_prompt]
        
        for i, col in enumerate(question_cols):
            if pd.notna(row[col]):  # Only include non-null answers
                prompt_parts.append(f"Q{i+1}: {col}\nA{i+1}: {row[col]}")
            else:
                # For missing answers, provide a placeholder
                prompt_parts.append(f"Q{i+1}: {col}\nA{i+1}: [No answer provided]")
        
        prompt_parts.append("Score:")
        prompt = "\n\n".join(prompt_parts)
        prompts.append(prompt)
    
    return prompts

# Modified scoring function to work with the original dataframe structure
def add_model_scores(df, question_cols, score_col, instruction_text, max_rows=None, batch_size=5):
    """
    Scores rows and adds the scores directly to the original dataframe.
    Returns the original dataframe with an additional model score column.
    """
    print(f"Processing scores for: {score_col}")
    
    # Create a copy to avoid modifying the original until we're ready
    df_copy = df.copy()
    
    # Create few-shot examples
    few_shot_prompt = create_few_shot_examples(df_copy, question_cols, score_col)
    full_prompt = few_shot_prompt + "\n\n---\n\n" + instruction_text
    
    # Determine which rows to score - focus on where at least first question is answered
    if len(question_cols) > 0 and len(df_copy) > 0:
        main_question = question_cols[0]  # Use the first question as the primary indicator
        rows_to_score = df_copy[df_copy[main_question].notna()].index.tolist()
    else:
        rows_to_score = df_copy.index.tolist()
    
    # Apply max_rows limit if specified
    if max_rows is not None and len(rows_to_score) > max_rows:
        rows_to_score = rows_to_score[:max_rows]
    
    if len(rows_to_score) == 0:
        print(f"No valid rows found for {score_col}")
        return df_copy
    
    print(f"Processing {len(rows_to_score)} rows")
    
    # Create a temporary dataframe with just the rows to score
    temp_df = df_copy.loc[rows_to_score].copy()
    
    # Generate prompts for scoring
    prompts = generate_prompts_for_scoring(temp_df, question_cols, full_prompt)
    
    # Score in batches
    model_scores = {}  # Use a dictionary with index as key
    
    # Simple counter to show progress
    total_batches = (len(prompts) + batch_size - 1) // batch_size
    
    for batch_idx in range(0, len(prompts), batch_size):
        current_batch = (batch_idx // batch_size) + 1
        print(f"Processing batch {current_batch}/{total_batches}")
        
        # Get indices and prompts for this batch
        batch_indices = rows_to_score[batch_idx:batch_idx + batch_size]
        batch_prompts = prompts[batch_idx:batch_idx + batch_size]
        
        for idx, prompt in zip(batch_indices, batch_prompts):
            try:
                response_text = generate(prompt)
                score = extract_score_from_output(response_text)
                
                if score is None:
                    print(f"Failed to extract score from: {response_text[:100]}...")
                
                model_scores[idx] = score
            except Exception as e:
                error_msg = str(e)
                print(f"Error: {error_msg}")
                model_scores[idx] = None
            
            # Add a small delay to avoid hitting rate limits
            time.sleep(0.5)
    
    # Add model scores directly to the original dataframe
    model_score_col = f"{score_col}_model"
    df_copy[model_score_col] = None
    
    for idx, score in model_scores.items():
        df_copy.at[idx, model_score_col] = score
    
    return df_copy

# Define instruction texts (unchanged)
instruction_texts = {
    "description": """
You are reviewing how educators use AI in their work. Your task is to evaluate the quality of their descriptions based on their answers to two questions:

1. How they use AI in their work
2. What educational challenge or goal it helps address

Rate the quality from 1 (poor) to 5 (excellent) based on:
- Clarity and specificity of the description
- Relevance to an educational challenge
- Thoughtfulness and innovation of the approach

After reviewing each example, provide a score from 1 to 5.
""",
    
    "evidence": """
You are reviewing evidence of improvements from educators using AI in their work. Your task is to evaluate the quality of their evidence based on their answers.

Rate the evidence from 1 (poor) to 5 (excellent) based on:
- Specificity and concreteness of the evidence provided
- Measurable or observable improvements mentioned
- Clarity of the connection between AI use and improvements

Note: Some responses may not have all fields filled in. Focus on evaluating the available information.

After reviewing each example, provide a score from 1 to 5.
""",
    
    "challenges": """
You are reviewing how educators have addressed challenges related to using AI in their work. Your task is to evaluate the quality of their explanations.

Rate the explanations from 1 (poor) to 5 (excellent) based on:
- Clarity of the challenge identification
- Effectiveness of the solution approach
- Thoughtfulness and creativity in addressing challenges

After reviewing each example, provide a score from 1 to 5.
""",
    
    "future": """
You are reviewing educators' future plans for using AI in their work. Your task is to evaluate the quality of their plans.

Rate the future plans from 1 (poor) to 5 (excellent) based on:
- Specificity and concreteness of the plans
- Alignment with educational goals
- Innovation and potential impact

After reviewing each example, provide a score from 1 to 5.
"""
}

# Set the maximum rows to process (set to None to process all)
max_rows_per_section = None  # For testing - set to None for full processing

# Process each dataframe with the improved function
# This function adds model scores directly to the original dataframes

# 1. Description
description_df = add_model_scores(
    description_df,
    ["How do you use AI in your work?", 
     "In one sentence, what educational challenge or goal does your AI use help you address?"],
    "Description of the practice (with examples)",
    instruction_texts["description"],
    max_rows=max_rows_per_section
)

# 2. Evidence - Modified to handle missing "If other, please specify:2" values
evidence_df = add_model_scores(
    evidence_df,
    ["Have you observed improvements as a result of using AI in your work?",
     "If possible, please provide evidence or examples of improvements"],  # Removed the "If other" field
    "Evidence of improvement and explanation",
    instruction_texts["evidence"],
    max_rows=max_rows_per_section
)

# 3. Challenges
challenges_df = add_model_scores(
    challenges_df,
    ["Did you manage to address those challenges? If so, please explain."],
    "Explanation on how they manage to address the challenges",
    instruction_texts["challenges"],
    max_rows=max_rows_per_section
)

# 4. Future
future_df = add_model_scores(
    future_df,
    ["What are your future plans for the use of this AI tool or AI-facilitated practice?"],
    "Explanation on what  are the future plans?",
    instruction_texts["future"],
    max_rows=max_rows_per_section
)

# Save the individual dataframes with model scores
description_df.to_excel("description_with_model_scores.xlsx", index=False)
evidence_df.to_excel("evidence_with_model_scores.xlsx", index=False)
challenges_df.to_excel("challenges_with_model_scores.xlsx", index=False)
future_df.to_excel("future_with_model_scores.xlsx", index=False)

print("All dataframes processed and saved with model scores.")

# If you want to create a consolidated dataframe, you can do so by joining on original_index
# Start with the original full dataframe
if 'df' in globals():  # Check if the original full df exists
    # Create a copy of the original full dataframe
    consolidated_df = df.copy()
    
    # Add model score columns from each section
    # Get model score columns
    description_model_col = "Description of the practice (with examples)_model"
    evidence_model_col = "Evidence of improvement and explanation_model"
    challenges_model_col = "Explanation on how they manage to address the challenges_model"
    future_model_col = "Explanation on what  are the future plans?_model"
    
    # Add model scores from description_df
    scores_from_description = description_df[["original_index", description_model_col]].set_index("original_index")
    consolidated_df = consolidated_df.join(scores_from_description, on="original_index")
    
    # Add model scores from evidence_df
    scores_from_evidence = evidence_df[["original_index", evidence_model_col]].set_index("original_index")
    consolidated_df = consolidated_df.join(scores_from_evidence, on="original_index")
    
    # Add model scores from challenges_df
    scores_from_challenges = challenges_df[["original_index", challenges_model_col]].set_index("original_index")
    consolidated_df = consolidated_df.join(scores_from_challenges, on="original_index")
    
    # Add model scores from future_df
    scores_from_future = future_df[["original_index", future_model_col]].set_index("original_index")
    consolidated_df = consolidated_df.join(scores_from_future, on="original_index")
    
    # Add overall average score
    consolidated_df["overall_model_score"] = consolidated_df[[
        description_model_col, 
        evidence_model_col, 
        challenges_model_col, 
        future_model_col
    ]].mean(axis=1)
    
    # Save the consolidated dataframe
    consolidated_df.to_excel("consolidated_survey_with_model_scores.xlsx", index=False)
    print("Consolidated dataframe created and saved.")
```
